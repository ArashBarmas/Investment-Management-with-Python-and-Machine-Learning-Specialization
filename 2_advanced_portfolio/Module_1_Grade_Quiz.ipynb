Load the 49 Value Weighted industry portfolio returns. Limit the analysis to returns of the 49 VW portfolios from 1991 onwards (1991 included, 1990 not included, 2018 included).
```python
y = ind49_rets.loc["1991-01":"2018-12", 'Beer'] - fff.loc["1991-01":"2018-12", "RF"]
X = fff.loc["1991-01":"2018-12", "Mkt-RF"]
X = X.to_frame()
X["alpha"] = 1

model = sm.OLS(y, X).fit()
model.summary()
```

### Question 6

Using the same data as the previous question, which of the 49 industries had the lowest CAPM Beta when evaluated over the 1991-1993 (both included) period? (Use the same industry names as in the files) . Enter the answer as text and remember to exactly match the column headers in the data file.

```python
betas = pd.Series(index = ind49_rets.columns, dtype=float)

for col in ind49_rets.columns:
    y = ind49_rets.loc['1991-01':'1993-12', col] - fff.loc['1991-01': '1993-12', "RF"]
    X = fff.loc['1991-01': '1993-12', "Mkt-RF"]
    X = X.to_frame()
    X["alpha"] = 1

    model = sm.OLS(y, X).fit()
    betas[col] = model.params["Mkt-RF"]
    

betas.sort_values()
```

```python
def capm_beta(r, start = '1991-01', end = '1993-12'):
    mkt_excess = fff.loc[start:end,['Mkt-RF']]
    exp_var = mkt_excess.copy()
    exp_var["Constant"] = 1
    
    if isinstance(r, pd.DataFrame):
        return r.aggregate(capm_beta)
    else:
        r_excess = pd.DataFrame(r[start:end]) - fff.loc[start:end, ['RF']].values
        lm = sm.OLS(r_excess, exp_var).fit()
        
        return lm.params[0]
```

```python
capm_beta(ind49_rets).sort_values()
```

### Question 8

Using the same dataset and period as the previous question, of the 49 industries, which displayed the highest Large Cap tilt when analyzed over the entire period?

```python
def fff_value(r, start = '1991', end = '2018'):
    mkt_excess = fff.loc[start:end,['Mkt-RF']]
    exp_var = mkt_excess.copy()
    exp_var["Value"] = fff.loc[start:end,['HML']]
    exp_var["Size"] = fff.loc[start:end,['SMB']]
    exp_var["Constant"] = 1
    
    if isinstance(r, pd.DataFrame):
        return r.aggregate(fff_value)
    else:
        r_excess = pd.DataFrame(r[start:end]) - fff.loc[start:end, ['RF']].values
        lm = sm.OLS(r_excess, exp_var).fit()
        
        return lm.params[1]
    
def fff_size(r, start = '1991', end = '2018'):
    mkt_excess = fff.loc[start:end,['Mkt-RF']]
    exp_var = mkt_excess.copy()
    exp_var["Value"] = fff.loc[start:end,['HML']]
    exp_var["Size"] = fff.loc[start:end,['SMB']]
    exp_var["Constant"] = 1
    
    if isinstance(r, pd.DataFrame):
        return r.aggregate(fff_size)
    else:
        r_excess = pd.DataFrame(r[start:end]) - fff.loc[start:end, ['RF']].values
        lm = sm.OLS(r_excess, exp_var).fit()
        
        return lm.params[2]
```
